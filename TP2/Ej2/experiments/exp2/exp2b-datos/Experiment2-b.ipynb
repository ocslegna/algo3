```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
df_500 = pd.read_csv('exp2nfijo.csv', delimiter=';')
```

```python
print("Parte b de la experimentación, n = 500, m variable")
ax = plt.gca()
for title, group in df_500.groupby('modo'):
    group = group.groupby('aristas').median().reset_index()
    group.plot(x='aristas', y='duracion', label=title, ax=ax)
plt.ylabel('ticks de reloj')
plt.xlabel('cantidad de vértices')

# Get current size
#fig_size = plt.rcParams["figure.figsize"]
#fig_size[0] = 10
#fig_size[1] = 8
#plt.rcParams["figure.figsize"] = fig_size

plt.axvline(x=95000, color = 'red', linestyle = ':')
plt.text(91000,20000000,'m = 95000',rotation=90):')
#plt.text(98500,8000000,'m = 97500',rotation=90)


plt.show()
```
